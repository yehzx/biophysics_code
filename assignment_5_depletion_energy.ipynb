### Problem
Please estimate the depletion interaction change based on the hyper/hypotonic conditions in the PNAS paper.

PNAS Paper:  
Orientation-independent-DIC imaging reveals that a transient rise in depletion attraction contributes to mitotic chromosome condensation. ([Paper Link](https://www.pnas.org/doi/10.1073/pnas.2403153121))

**AO depletion potential (hard sphere):**
$$U_{sphere}(h) = -\frac{\pi}{2} k_B T n_s a \sigma^2 (1 - \frac{h}{\sigma})^2,$$ 
where $a$ is the radius of the large sphere, $\sigma$ is the diameter of the small sphere, and $n_s = \frac{\phi}{\frac{4}{3} \pi \sigma^3}$ is the number density of the small sphere ($\phi$ is the volume fraction). 




### Approach

To calculate the depletion interaction energy, we need to the chromosome milieu (cytoplasm) density under hyper/hypotonic conditions. This information is provided in Figure 4 and 5. The density is 238 mg/mL under hypertonic conditions and 93 mg/mL under hypotonic conditions (<10 min after treatment) in HCT116 cells. The density is 164 mg/mL under normal condiditons.  

Let's assume the depletant molecules are proteins of about 30kDa with a diameter of 3 nm, as mentioned in the paper, and a chromatin domain with a radius of 100 nm. The depletion interaction energy is given by:

(1) Hypertonic
$$\frac{U_{sphere}(h)}{k_B T} = -\frac{\pi}{2} ((\frac{\color{orange}{238} \times 10^{-27} kg \cdot nm^{-3}}{30 kg \cdot mol^{-1}}) (\frac{6.022 \times 10^{23}}{1 mol})) (100 nm) (3 nm)^2 (1 - \frac{h}{3 nm})^2 $$

(2) Hypotonic
$$\frac{U_{sphere}(h)}{k_B T} = -\frac{\pi}{2} ((\frac{\color{orange}{93} \times 10^{-27} kg \cdot nm^{-3}}{30 kg \cdot mol^{-1}}) (\frac{6.022 \times 10^{23}}{1 mol})) (100 nm) (3 nm)^2 (1 - \frac{h}{3 nm})^2 $$



```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_u(h, d):
    return - np.pi / 2 * (d * 10**-27 / 30) * (6.022 * 10**23) * 100 * (3**2) * ((1 - h / 3)**2)

x = np.arange(0, 3, 0.1)
plt.plot(x, calculate_u(x, 238), label="hypertonic")
plt.plot(x, calculate_u(x, 93), label="hypotonic")
plt.ylabel(r"Potential ($k_B T$)")
plt.xlabel("Separation (h) (nm)")
plt.legend()
```
